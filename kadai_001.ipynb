```python
import openpyxl

workbook = openpyxl.Workbook()
sheet = workbook.active

sheet['A2'].value = '請求書'
sheet['A4'].value = '株式会社ABC'
sheet['A5'].value = '〒101-0022 東京都千代田区神田練堀町300'
sheet['A6'].value = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['A7'].value = '担当者名:鈴木一郎 様'

sheet['E4'].value = 'No.'
sheet['F4'].value = '0001'
sheet['E5'].value = '日付'

from datetime import datetime
today = datetime.today()

sheet['F5'].value = today.strftime('%Y/%m/%d')

header = ['商品名', '数量', '単価', '金額']
sheet.append(header)

data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000],
]

for row in data:
    sheet.append(row)

sheet.insert_rows(8, 2)

sheet['D13'].value = '=SUM(E11:E12)'

sheet['A15'].value = '合計'
sheet['A16'].value = '消費税'
sheet['A17'].value = '税込合計'

sheet['D15'].value = '=E13'
sheet['D16'].value = '=E15*0.1'
sheet['D17'].value = '=SUM(E15:E16)'

sheet.insert_cols(1, 1)

filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"

workbook.save(filename)
```
